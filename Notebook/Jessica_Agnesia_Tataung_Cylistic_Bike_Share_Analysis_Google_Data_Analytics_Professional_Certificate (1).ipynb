```python
# Import Library
import pandas as pd
import numpy as np
```

```python
# Load data
df_2019 = pd.read_csv('/content/Divvy_Trips_2019_Q1 - Divvy_Trips_2019_Q1.csv')
df_2020 = pd.read_csv('/content/Divvy_Trips_2020_Q1 - Divvy_Trips_2020_Q1.csv')
```

```python
# Displaying the first 5 rows of 2019 data
df_2019.head()
```

```python
# Displaying the first 5 rows of 2020 data
df_2020.head()
```

```python
# Rename columns
df_2019 = df_2019.rename(columns={
    'start_time': 'started_at',
    'end_time': 'ended_at',
    'from_station_name': 'start_station_name',
    'from_station_id': 'start_station_id',
    'to_station_name': 'end_station_name',
    'to_station_id': 'end_station_id',
    'usertype': 'member_casual'
})
```

```python
# Combine data
df = pd.concat([df_2019, df_2020])
```

```python
# Displaying the merged dataframe
df
```

```python
# Make tripduration column
df['tripduration'] = (
    pd.to_datetime(df['ended_at']) -
    pd.to_datetime(df['started_at'])
).dt.total_seconds()
```

```python
# Change subcriber and customer to member and casual
df['member_casual'] = df['member_casual'].replace({
    'Subscriber': 'member',
    'Customer': 'casual'
})
```

```python
df
```

```python
# Check the data types, column counts, and null distributions across the combined dataframe
df.info()
```

```python
# Change datatype float --> int
df['trip_id'] = df['trip_id'].astype('Int64')
df['bikeid'] =df['bikeid'].astype('Int64')
df['start_station_id'] = df['start_station_id'].astype('Int64')
df['end_station_id']=df['end_station_id'].astype('Int64')
```

```python
# Change data type --> object
df['trip_id'] = df['trip_id'].astype('object')
df['bikeid'] =df['bikeid'].astype('object')
df['start_station_id'] = df['start_station_id'].astype('object')
df['end_station_id']=df['end_station_id'].astype('object')
df['ride_id']= df['ride_id'].astype('object')
```

```python
# Change data type --> Category
df['member_casual'] = df['member_casual'].astype('category')
df['gender'] = df['gender'].astype('category')
df['rideable_type'] = df['rideable_type'].astype('category')
```

```python
# Change data type object --> int
df['tripduration'] = df['tripduration'].astype('Int64')
df['birthyear'] = df['birthyear'].astype('Int64')
```

```python
# Change data type --> datetime
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at']= pd.to_datetime(df['ended_at'])
```

```python
# Make day of week column
df['day_of_week'] = df['started_at'].dt.day_name()
```

```python
# Coalesce unaligned trip identifiers into a single unified column
df['unified_trip_id'] = df['ride_id'].fillna(df['trip_id'])
```

```python
# Drop original fragmented ID columns to finalize data structural cleaning
df.drop(['ride_id', 'trip_id'], axis=1, inplace=True)
```

```python
# Verify column data types and check for missing values after cleaning
df.info()
```

```python
# Generate descriptive stats to spot outliers
df.describe()
```

```python
# Check duplicated
df.duplicated().sum()
```

```python
# Missing value
missing = df.isna().sum()
missing_percent = (df.isna().sum() / len(df)) * 100
```

```python
# Create a summary dataframe to audit missing value counts and percentages
missing_df = pd.DataFrame({
    'Missing Value' : missing,
    'Missing Value (%)' : missing_percent
})

missing_df
```

The high missing percentages (~46% and ~56%) in columns like gender, birthyear, and GPS coordinates are due to structural schema differences between the 2019 and 2020 datasets. Since dropping rows would accidentally delete an entire year of data, these inconsistent columns will simply be excluded from the final analysis to maintain data integrity

```python
# detail of missing value
df[df['end_station_name'].isna()]
```

```python
# drop missing value
df.dropna(subset=['end_station_name'], inplace=True)
```

```python
# Recheck missing value
missing = df.isna().sum()

missing_percent = (df.isna().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Value': missing,
    'Missing Value (%)': missing_percent
})

missing_df
```

```python
# Preview trips that under 0 second (potentially faulty data)
df[df['tripduration']<0]
```

```python
# Filter out the 117 false trips from the main dataframe
df = df[~(df['tripduration'] < 0)]
```

```python
# Preview trips that start & end at the same station under 1 minute/60 seconds (potentially faulty data)
df[(df['start_station_name']==df['end_station_name']) & (df['tripduration'] <= 60) ]
```

an evaluation of short-duration trips revealed 7,452 rows where the ride started and ended at the exact same station (start_station_name == end_station_name) with a total duration of 1 minute or less (tripduration <= 60). While returning to the same station is normal for leisure, sightseeing, or short exercise, a duration under 60 seconds strongly indicates false starts, docking issues, or immediate cancellations due to faulty equipment

```python
# Filter out the 7,568 false trips from the main dataframe
df = df[~(
    (df['start_station_name'] == df['end_station_name']) &
    (df['tripduration'] <= 60)
)]
```
